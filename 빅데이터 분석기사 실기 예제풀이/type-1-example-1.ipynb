### mrcars 데이터셋의 qsec 컬럼을 최소최대척도(Min-Max Scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수를 구하시오.

### DataUrl = https://raw.githubusercontent.com/Datamanim/dataq/main/mtcars.csv

```python
import pandas as pd
DataUrl = 'https://raw.githubusercontent.com/Datamanim/dataq/main/mtcars.csv'
df = pd.read_csv(DataUrl)
```

```python
df
```

```python
df['qsec']
```

```python
from sklearn.preprocessing import MinMaxScaler
'''
Examples
 |  --------
 |  >>> from sklearn.preprocessing import MinMaxScaler
 |  >>> data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
 |  >>> scaler = MinMaxScaler()
 |  >>> print(scaler.fit(data))
 |  MinMaxScaler()
 |  >>> print(scaler.data_max_)
 |  [ 1. 18.]
 |  >>> print(scaler.transform(data))
 |  [[0.   0.  ]
 |   [0.25 0.25]
 |   [0.5  0.5 ]
 |   [1.   1.  ]]
 |  >>> print(scaler.transform([[2, 2]]))
 |  [[1.5 0. ]]
 '''
```

```python
data = df['qsec'].values.reshape(-1,1)
data
```

```python
scaler = MinMaxScaler()
scaler.fit(data)
y = scaler.transform(data)
y
```

```python
print(y>0.5)
```

```python
print((y>0.5).sum())
```

### mrcars 데이터셋의 qsec 컬럼을 최소최대척도(Min-Max Scale)로 변환한 후 0.5보다 큰 값을 가지는 레코드 수는 9개
